```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score 
import pickle
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
X_train,X_test,y_train,y_test=train_test_split(independent,dependent, test_size=0.30, random_state=0)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
y_train
```

```python
regressor=SVR()
regressor.fit(X_train,y_train)
```

```python
nsupport=regressor.n_support_
nsupport
```

```python
support=regressor.support_
support
```

```python
bias = regressor.intercept_
bias
```

```python
y_pred=regressor.predict(X_test)
```

```python
r_score=r2_score(y_test,y_pred)
r_score
```

```python
file_name="finalized_svm_model.sav"
file_name_sc_x="svr_model_sc_x.sav"
file_name_sc_y="svr_model_sc_y.sav"
pickle.dump(regressor,open(file_name,"wb"))
pickle.dump(sc,open(file_name_sc_x,"wb"))
pickle.dump(scy,open(file_name_sc_y,"wb"))
```

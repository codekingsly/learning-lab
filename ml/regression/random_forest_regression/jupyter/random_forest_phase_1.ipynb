```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import pickle
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
regressor=RandomForestRegressor()
regressor.fit(X_train,y_train)
```

```python
y_predict=regressor.predict(X_test)
```

```python
rscore=r2_score(y_test,y_predict)
rscore
```

```python
file_name="finalized_random_forest_model.sav"
pickle.dump(regressor,open(file_name,'wb'))
```

```python
load_model=pickle.load(open("finalized_random_forest_model.sav","rb"))
result=load_model.predict([[38, 27, 2, True, False]])
result
```
